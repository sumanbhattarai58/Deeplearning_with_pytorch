```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
#Load datasets
data = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
data.head()
```

```python
#see the datatype of data
data.info()
```

```python
#drop columns 'id' and 'unnamed'
data.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
#check the columns and shape
print(data.columns)
print(data.shape)
```

```python
data.head()
```

```python
data['diagnosis'].unique()
```

```python
#train test split (assign column 0 for y, and rest for x)
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, 1:], data.iloc[:, 0], test_size=0.2, random_state=0)
```

```python
#scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
y_train
```

```python
#label encoding for y
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

```python
X_train.dtype
```

```python
#numpy arrays to pytorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32)
```

```python
y_train_tensor.dtype
```

```python
print(X_train_tensor.shape)
print(X_test_tensor.shape[1])
print(y_train_tensor.shape)
```

```python
print(X_test_tensor.shape)
print(y_test_tensor.shape)
```

```python
#Defining the model
import torch.nn as nn

class BreastCancer(nn.Module):

  def __init__(self,input_features):
    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(input_features,5),
        nn.ReLU(),
        nn.Linear(5,1),
        nn.Sigmoid()
    )

  def forward(self,X):
    out = self.network(X)
    return out
```

Important parameters

```python
learning_rate = 0.1
epochs = 50
```

```python
#Define loss function
loss_fn = nn.BCELoss()
```

Training Pipeline

```python
#create object
model =BreastCancer(X_train_tensor.shape[1])

#define optimizer
optimizer = torch.optim.SGD(model.parameters(),lr = learning_rate)

#define loop
for epoch in range(epochs):

  #forward pass
  y_pred =model(X_train_tensor)

  #calculate loss
  loss = loss_fn(y_pred,y_train_tensor.unsqueeze(1))

  #backward pass
  loss.backward()

  #update weights
  optimizer.step()

  #zero gradients
  optimizer.zero_grad()

  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

Evaluation

```python
#model evaluation
with torch.no_grad():   #temporarily disable gradient tracking
  y_pred = model.forward(X_test_tensor)
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_test_tensor.unsqueeze(1)).float().mean() #adds a new dimension [batch] into [batch_size, 1]).
  print(f'Accuracy: {accuracy.item()}')
```

Using Dataset and DataLoader

```python
from torch.utils.data import Dataset, DataLoader

class customdataset(Dataset):

  def __init__(self,features, labels):
    self.features = features
    self.labels = labels

  def __len__(self):
      return len(self.features)

  def __getitem__(self, idx):
      return self.features[idx], self.labels[idx]
```

```python
train_dataset = customdataset(X_train_tensor, y_train_tensor)
test_dataset = customdataset(X_test_tensor, y_test_tensor)
```

```python
train_dataset[10]
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

Training Pipeline

```python
#define loop
for epoch in range(epochs):
  for batch_features, batch_labels in train_loader:

    #forward pass
    y_pred1 = model(batch_features)

    #calculate loss
    loss1 = loss_fn(y_pred1, batch_labels.unsqueeze(1))

    #backward pass
    loss1.backward()

    #update weights
    optimizer.step()

    #zero gradients
    optimizer.zero_grad()

  # print loss in each epoch
    print(f'Epoch: {epoch + 1}, Loss: {loss1.item()}')
```

Evaluation

```python
# Model evaluation using test_loader
model.eval()  # Set the model to evaluation mode
accuracy_list = []

with torch.no_grad():
    for batch_features, batch_labels in test_loader:
        # Forward pass
        y_pred1 = model(batch_features)
        y_pred1 = (y_pred1 > 0.5).float()  # Convert probabilities to binary predictions

        # Calculate accuracy for the current batch
        batch_accuracy = (y_pred1.view(-1) == batch_labels).float().mean().item()
        accuracy_list.append(batch_accuracy)

# Calculate overall accuracy
overall_accuracy = sum(accuracy_list) / len(accuracy_list)
print(f'Accuracy: {overall_accuracy:.4f}')
```
